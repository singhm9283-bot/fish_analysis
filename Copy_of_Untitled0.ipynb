```python
import pandas as pd
import matplotlib.pyplot as plt


df=pd.read_csv('Fish.csv')
print(df.head())



```

```python
x = df['Weight']
y = df['Height']
plt.scatter(x,y)



```

```python
species_count=df['Species'].value_counts()
print(species_count)
```

```python
plt.pie(species_count,labels=species_count.index)
plt.title('Fish species')
```
